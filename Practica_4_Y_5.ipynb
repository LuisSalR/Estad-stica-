```python
import random
edades = []
for i in range(100):
  numero_aleatorio = random.randint( 18,70)
  edades.append(numero_aleatorio)


print(edades)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(edades,columns=["Edad"])
df.head()
```

```python
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad','Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

print("Tabla de Distribucion de frecuencias")
tabla_frecuencias

```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia']/100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec porcentual'] = tabla_frecuencias['Frecuencia']/100*100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias




```

#Regla de Sturges
1. Es una regla que sirve para calcular el numero de clases o intervalos idoneo en los que se debe dividir un conjunto de datos
2. Establece que el numero de clases es igual a uno más el logaritmo en base dos del numero total de datos








```python
import math
```

```python
print(edades)
```

```python
df = pd.DataFrame({'Edad': edades})
df


```

```python
n = len(edades)
K = int( 1 + math.log2(100))
print("Valor de n =", n,"\nNúmero de intervalos =", K)
```

```python
tabla_frecuencia = pd.cut( df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frec']
tabla_frecuencia
```

```python
tabla_frecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
import matplotlib.pyplot as plt
plt.hist(edades,alpha = 0.5, bins =15)
plt.title("Histograma de edades basado en la funcion de matplotlib con valores estandar")





```

```python
plt.hist(df['Edad'], bins=K, alpha = 0.5)
plt.title('Histograma de edades basado en la regla de Sturges')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')




```

#PRACTICA 5 DIAGRAMA DE CAJAS Y BIGOTES

```python
import random
# lista, vector que representa la edad de 100 personas
edades2 =[]
# ciclo  para generar numeros aleatorios
for i in range (100):
  #generar el numero aleatorio con la funcion randit y se guarda en la variable numero_aleatorio
  numero_aleatorio= random.randint(20, 87)
# Agreganos el numero aleatorio a la lista de valores
  edades2.append(numero_aleatorio)
print (edades2)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9,4))
bgplot1 = ax1.boxplot(edades, vert=True, patch_artist=True)
ax1.set_title('Edades encuesta 1')
bgplot2 = ax2.boxplot(edades2, vert=True, patch_artist=True)
ax2.set_title('Edades Encuesta 2')
plt.show()
```
